リストにするとエラーになる　numpyのarray配列にすることで四則演算が容易に出来る

```python
x = [1,2,3,4,5]
y = np.array([1,4,9,16,24])
plt.plot(x,y)
plt.plot(x,y/2)
plt.plot(x,y/3)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('title')
plt.show()
```

それぞれｘとｙに数値を配列　np.array配列にしてｙ軸１倍２分の１倍３分の１倍３本の線を表示した場合

```python
x = [1,2,3,4,5]
y = np.array([1,4,9,16,24])
plt.plot(x,y,color='r', label='A')
plt.plot(x,y/2,color='b',label='B')
plt.plot(x,y/3,color='g',label='C')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('title')
plt.legend()
plt.show()
```

それぞれの線にlabel='名前'を決めて 凡例を挿入　plt.legend()

```python
x = [1,2,3,4,5]
y = np.array([1,4,9,16,24])
plt.plot(x,y,color='r', label='A')
plt.plot(x,y/2,color='b',label='B')
plt.plot(x,y/3,color='g',label='C')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('title')
plt.legend(fontsize=15)
plt.show()
```
