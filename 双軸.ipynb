```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# ヘッダー行がなく、カラム名を指定したい場合
df1 = pd.read_csv('usd.jpy_twin - シート1.csv', parse_dates=[0], index_col=0, header=None, names=['Date', 'USD/JPY'])

# ヒストグラムの作成
plt.hist(df1['USD/JPY'], bins=30, density=True)

# 平均と標準偏差を取得して、正規分布を近似させる
mu = np.mean(df1['USD/JPY'])
sigma = np.std(df1['USD/JPY'])
x = np.linspace(df1['USD/JPY'].min(), df1['USD/JPY'].max(), len(df1['USD/JPY']))
p = stats.norm.pdf(x, mu, sigma)

# 確率密度関数をプロットする
plt.plot(x, p, 'k', linewidth=2)

# グラフの表示
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルからデータを読み込む
df = pd.read_csv("price_530 - シート1.csv")

# データフレームのカラム名を指定する
df.columns = ['Date','USDJPY', 'GBPJPY','EURJPY','EURUSD','GBPUSD']
# データフレームのカラム名を絞る
#df.columns = ['Date','USDJPY', 'GBPUSD']
# それぞれ変化率を求めてカラムを作成

# 変化率の数値ラインを決めて以上と以下のカラムを作成
```

```python
# それぞれ変化率を求めてカラムを作成
df['USDJPY_change'] = df['USDJPY'].pct_change()
df['GBPUSD_change'] = df['GBPUSD'].pct_change()

# 変化率の数値ラインを決める（例：10pips）
threshold = 0.001

# 変化率が数値ライン以上の場合は1、-数値ライン以下の場合は-1、それ以外の場合は0とするカラムを作成
df['USDJPY_label'] = df['USDJPY_change'].apply(lambda x: 1 if x >= threshold else (-1 if x <= -threshold else 0))
df['GBPUSD_label'] = df['GBPUSD_change'].apply(lambda x: 1 if x >= threshold else (-1 if x <= -threshold else 0))

# 1日括りで作成することで週ごとの特徴　日別　時間別の特徴等を抽出
# 日付をdatetime型に変換
df['Date'] = pd.to_datetime(df['Date'])

# 日付をインデックスに設定
df.set_index('Date', inplace=True)

# 1日ごとにデータを集約する（例：最終値）
df_daily = df.resample('D').last()

# データを表示する
print(df_daily)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルからデータを読み込む
df = pd.read_csv("prtice_53031.csv")

# データフレームのカラム名を指定する
df.columns = ['Date','USDJPY', 'GBPJPY','EURJPY','EURUSD','GBPUSD']

# プロットのサイズを指定する
fig, ax = plt.subplots(figsize=(10, 8))

# USD/JPYの折れ線グラフを描く
ax.plot(df.index, df['USDJPY'], color='b')
ax.set_xlabel('Date')
ax.set_ylabel('USDJPY', color='b')
ax.tick_params('y', colors='b')

# GBP/USDの折れ線グラフを描く
ax2 = ax.twinx()
ax2.plot(df.index, df['GBPUSD'], color='r')
ax2.set_ylabel('GBPUSD', color='r')
ax2.tick_params('y', colors='r')

# プロットを表示する
plt.show()

# それぞれ前の行との差分を求めてカラムを作成
df['USDJPY_diff'] = df['USDJPY'].diff()
df['GBPUSD_diff'] = df['GBPUSD'].diff()

# 差分カラム同士の積を計算してカラムを作成
df['product'] = df['USDJPY_diff'] * df['GBPUSD_diff']

# 積カラムが負である行だけを選択
df_negative = df[df['product'] < 0]

# データを表示する
print(df_negative)
```
