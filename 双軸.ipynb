```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ヘッダー行がなく、カラム名を指定したい場合
df1 = pd.read_csv('usd.jpy_twin - シート1.csv', parse_dates=[0], index_col=0, header=None, names=['Date', 'USD/JPY'])
df2 = pd.read_csv('gbp.usd_twin - シート1.csv', parse_dates=[0], index_col=0, header=None, names=['Date', 'GBP/USD'])
```

```python
fig, ax = plt.subplots()
ax.plot(df1.index, df1['USD/JPY'], color='b')
ax.set_xlabel('Date')
ax.set_ylabel('USD/JPY', color='b')
ax.tick_params('y', colors='b')
ax2 = ax.twinx()
ax2.plot(df2.index, df2['GBP/USD'], color='r')
ax2.set_ylabel('GBP/USD', color='r')
ax2.tick_params('y', colors='r')
plt.show()
```

```python
# 2軸のグラフを作成する
fig, ax = plt.subplots()
plt.xticks(rotation=90)
ax.plot(df1.index, df1['value01'], color='b')
ax.set_xlabel('Data')
ax.set_ylabel('Value01', color='b')
ax.tick_params('y', colors='b')
ax2 = ax.twinx()
ax2.plot(df2.index, df2['value02'], color='r')
ax2.set_ylabel('Value02', color='r')
ax2.tick_params('y', colors='r')
plt.show()
```
